```python
# Importar las bibliotecas necesarias
import requests
import zipfile
import os
import datetime
import pandas as pd
import numpy as np
```

```python
# Definir la URL del archivo ZIP a descargar
url = "http://mipronostico.ideam.gov.co/IdeamWebApp2/Ideam/getDatosAbiertos/pronosticos/ws/pronCsv.zip"

# Realizar la solicitud HTTP para descargar el archivo ZIP
response = requests.get(url)

# Verificar si la descarga fue exitosa (código de estado HTTP 200)
if response.status_code == 200:
    # Abrir el archivo ZIP descargado en modo binario para escritura
    with open("downloaded_file.zip", "wb") as file:
        file.write(response.content)

    # Imprimir un mensaje de éxito
    print("Archivo ZIP descargado exitosamente.")

    # Descomprimir el archivo ZIP y conservar solo el CSV
    with zipfile.ZipFile("downloaded_file.zip", "r") as zip_ref:
        # Obtener una lista de archivos en el archivo ZIP que tengan extensión .csv
        csv_files = [name for name in zip_ref.namelist() if name.lower().endswith('.csv')]

        # Verificar si se encontraron archivos CSV en el archivo ZIP
        if csv_files:
            # Tomar el primer archivo CSV encontrado
            csv_filename = csv_files[0]
            csv_basename, csv_extension = os.path.splitext(csv_filename)
            current_date = datetime.datetime.now().strftime("%Y-%m-%d")

            # Crear un nuevo nombre para el archivo CSV con la fecha actual
            new_csv_filename = f"{current_date}_{'Pronostico_Del_Tiempo'}_{csv_basename}{csv_extension}"

            # Extraer el archivo CSV del ZIP
            zip_ref.extract(csv_filename)

            # Renombrar el archivo CSV extraído con el nuevo nombre
            os.rename(csv_filename, new_csv_filename)

            # Eliminar el archivo ZIP descargado
            os.remove("downloaded_file.zip")

            # Imprimir un mensaje con el nombre del archivo CSV renombrado
            print(f"Archivo CSV extraído y renombrado como: {new_csv_filename}")
        else:
            # Si no se encontraron archivos CSV en el ZIP
            print("No se encontraron archivos CSV en el archivo ZIP.")
else:
    # Si hubo un error al descargar el archivo ZIP
    print(f"Error al descargar el archivo ZIP. Código de estado: {response.status_code}")
```

```python
import requests
import boto3
import io
import zipfile
import os
import datetime
import tempfile

def lambda_handler(event, context):
    try:
        # Definir la URL del archivo ZIP a descargar
        url = "http://mipronostico.ideam.gov.co/IdeamWebApp2/Ideam/getDatosAbiertos/pronosticos/ws/pronCsv.zip"

        # Realizar la solicitud HTTP para descargar el archivo ZIP
        response = requests.get(url)

        # Verificar si la descarga fue exitosa (código de estado HTTP 200)
        if response.status_code == 200:
            # Crear un directorio temporal
            with tempfile.TemporaryDirectory() as temp_dir:
                # Crear un cliente de S3
                s3_client = boto3.client("s3")

                # Nombre del bucket de S3
                bucket_name = "colombiaweather"

                # Nombre de la carpeta en S3 donde guardar los archivos
                s3_folder = "Colombia_Weather_Ideam"

                # Crear un objeto BytesIO para almacenar los datos del archivo ZIP
                zip_data = io.BytesIO(response.content)

                # Extraer el archivo ZIP
                with zipfile.ZipFile(zip_data, "r") as zip_ref:
                    # Obtener una lista de archivos en el archivo ZIP que tengan extensión .csv
                    csv_files = [name for name in zip_ref.namelist() if name.lower().endswith('.csv')]

                    if csv_files:
                        csv_filename = csv_files[0]
                        csv_basename, csv_extension = os.path.splitext(csv_filename)
                        current_date = datetime.datetime.now().strftime("%Y-%m-%d")
                        new_csv_filename = f"{current_date}_Pronostico_Del_Tiempo{csv_extension}"

                        # Extraer el archivo CSV del ZIP
                        zip_ref.extract(csv_filename, path=temp_dir)

                        # Renombrar el archivo CSV extraído con el nuevo nombre
                        os.rename(os.path.join(temp_dir, csv_filename), os.path.join(temp_dir, new_csv_filename))

                        # Subir el archivo CSV renombrado a S3
                        s3_filename = f"{s3_folder}/{new_csv_filename}"
                        s3_client.upload_file(os.path.join(temp_dir, new_csv_filename), bucket_name, s3_filename)

                        # Eliminar el ZIP descargado
                        # os.remove("pronCsv.zip")

                        return "Archivo ZIP descargado, CSV extraído y guardado en S3 exitosamente."
                    else:
                        return "No se encontraron archivos CSV en el archivo ZIP."
        else:
            return f"Error al descargar el archivo ZIP. Código de estado: {response.status_code}"
    except Exception as e:
        return f"Ocurrió un error: {str(e)}"
```
